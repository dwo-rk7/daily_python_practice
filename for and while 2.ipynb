```python
#Write a program to print numbers from 1 to 20 using a while loop.
num=1
while num<=20:
    print(num)
    num=num+1
```

```python
#Write a program to print all even numbers between 1 and 50 using a while loop.
num1=1
while num1<=50:
    if num1%2==0:
        print(" even num ",num1)
    num1=num1+1
```

```python
#Write a program to calculate the sum of first n natural numbers.
summ=0
n=1
while n<=15:
    summ=n+summ
    n=n+1
    
print(summ)
```

```python
#Write a program to reverse a number using a while loop.
no=int(input("enter the number"))
rev=0
while no>0:
    digit=no%10
    rev=rev*10+digit
    no=no//10
print(rev)
```

```python
#Write a program to check whether a number is palindrome or not.
num=int(input("enter the number"))
ori=num
rev=0
while num>00:
    digit=num%10
    rev=rev*10+digit
    num=num//10
    

if ori==rev:
    print("palindrome")
else:
    print("not a palindrome number")
```
