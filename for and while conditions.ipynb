```python
#15. Take a positive number and keep dividing it by 2 using while loop until it becomes less
# than 1. Count the steps.
number=int(input("enter how many numbers you want to divide"))
result=float(input("enter the number 1 "))
count=1 

while count<number:
    num=float(input(f"enter the number {count+1}"))
    if num==0:
        print("zero division error")
        break
    result /= num
    count += 1  
print(f"quotient of {num} numbers {result}")
```

```python
#16. Use while loop to reverse a number (e.g., 1234 → 4321).

digit=int(input("enter the numnber"))
rev=0
while digit>0:
    num=digit%10
    rev=rev*10+num
    digit=digit//10
print(rev)
```

```python
#19. Use while loop to check if a given number is prime or not.
num = int(input("Enter a number: "))

i = 2
is_prime = True

while i < num:
    if num % i == 0:
        is_prime = False
        break
    i += 1

if num <= 1:
    print(num, "is not a Prime number")
elif is_prime:
    print(num, "is a Prime number")
else:
    print(num, "is not a Prime number")
```
