```python
#23.Take 4 student names and their marks as input from the user. store them in a dictionary (name as key, marks as value). Then find and
# print the student with the highest marks.
student = {}

for i in range(4):
    name = input("Enter the name : ")
    marks= input("Enter the Marks :")
    student[name] = marks

print(student)
highest_marks= max(student,key =student.get)

highest_marks_std =student[highest_marks]

print(f"Highest marks students name is {highest_marks} and marks {highest_marks_std} ")
```
