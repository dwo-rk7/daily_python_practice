```python
#primr number question


for num in range(11,51):
   
    if num>1:

        for i in range(2,num):
            
            if num%i==0:
                print(f"{num} : not prime")
                break
            else:
                print(f"{num} : prime")
                
        
    else:
        print("not prime")
        
    num+=1
```
